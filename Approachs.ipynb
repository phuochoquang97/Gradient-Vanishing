```python
# fashion_mnist
import tensorflow as tf
import tensorflow.keras as keras
import numpy as np

(X_train, y_train), (X_test, y_test) = keras.datasets.fashion_mnist.load_data()
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
X_train = X_train.reshape(60000, -1)
X_test  = X_test.reshape(10000, -1)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# METHOD 1: USING ADAM OPTIMIZER

```python
# set seed
tf.random.set_seed(1234)
initializer = tf.keras.initializers.RandomNormal()

# create model
model = keras.Sequential()
model.add(keras.Input(shape=(784,)))
model.add(keras.layers.Dense(128, activation='sigmoid',
                             kernel_initializer=initializer))
model.add(keras.layers.Dense(128, activation='sigmoid',
                             kernel_initializer=initializer))
model.add(keras.layers.Dense(128, activation='sigmoid',
                             kernel_initializer=initializer))
model.add(keras.layers.Dense(128, activation='sigmoid',
                             kernel_initializer=initializer))
model.add(keras.layers.Dense(128, activation='sigmoid',
                             kernel_initializer=initializer))
model.add(keras.layers.Dense(128, activation='sigmoid',
                             kernel_initializer=initializer))
model.add(keras.layers.Dense(128, activation='sigmoid',
                             kernel_initializer=initializer))
model.add(keras.layers.Dense(10, activation='softmax'))


# optimizer and loss
model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy', 
              metrics=['sparse_categorical_accuracy'])

# training
batch_size = 512
history = model.fit(X_train, y_train, batch_size,
                    validation_data=(X_test, y_test),
                    epochs=100, verbose=2)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.xlabel('iteration')
plt.ylabel('Loss')
plt.legend()
```

```python
 plt.plot(history.history['sparse_categorical_accuracy'], label='train_accuracy')
plt.plot(history.history['val_sparse_categorical_accuracy'], label='val_accuracy')
plt.xlabel('iteration')
plt.ylabel('Accuracy')
plt.legend()
```

# METHOD 4: USING RESIDUAL NETWORKS

```python
# problem
import tensorflow as tf
import tensorflow.keras as keras

# set seed
tf.random.set_seed(1234)
initializer = tf.keras.initializers.RandomNormal()


######################
model = keras.Sequential()
ip = keras.Input(shape=(784,))

layer1 = keras.layers.Dense(128, activation='sigmoid', kernel_initializer=initializer)
x1 = layer1(ip)

layer2 = keras.layers.Dense(128, activation='sigmoid', kernel_initializer=initializer)
x2 = layer2(x1)

layer3_dense = keras.layers.Dense(128, kernel_initializer=initializer)
x3_dense = layer3_dense(x2)
x_add = keras.layers.Add()([x3_dense, x1])
x3 = keras.activations.sigmoid(x_add)

layer4_dense = keras.layers.Dense(128, kernel_initializer=initializer)
x4_dense = layer4_dense(x3)
x_add = keras.layers.Add()([x4_dense, x2])
x4 = keras.activations.sigmoid(x_add)

layer5_dense = keras.layers.Dense(128, kernel_initializer=initializer)
x5_dense = layer5_dense(x4)
x_add = keras.layers.Add()([x5_dense, x3])
x5 = keras.activations.sigmoid(x_add)

layer6_dense = keras.layers.Dense(128, kernel_initializer=initializer)
x6_dense = layer6_dense(x5)
x_add = keras.layers.Add()([x6_dense, x4])
x6 = keras.activations.sigmoid(x_add)

layer7_dense = keras.layers.Dense(128, kernel_initializer=initializer)
x7_dense = layer7_dense(x6)
x_add = keras.layers.Add()([x7_dense, x5])
x7 = keras.activations.sigmoid(x_add)

op = keras.layers.Dense(10, activation='softmax')(x7)

model = keras.models.Model(ip, op)


# optimizer and loss
model.compile(optimizer='sgd', 
              loss='sparse_categorical_crossentropy', 
              metrics=['sparse_categorical_accuracy'])

# training
batch_size = 512
history = model.fit(X_train, y_train, batch_size,
                    validation_data=(X_test, y_test),
                    epochs=100, verbose=2)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.xlabel('iteration')
plt.ylabel('Loss')
plt.legend()
```

```python
plt.plot(history.history['sparse_categorical_accuracy'], label='train_accuracy')
plt.plot(history.history['val_sparse_categorical_accuracy'], label='val_accuracy')
plt.xlabel('iteration')
plt.ylabel('Accuracy')
plt.legend()
```

```python
class MySigmoidActivation(tf.keras.layers.Layer):
    
    def __init__(self):
        super(MySigmoidActivation, self).__init__()
        
    def call(self, inputs):
        return tf.keras.activations.sigmoid(inputs) + 0.15 * inputs
```
